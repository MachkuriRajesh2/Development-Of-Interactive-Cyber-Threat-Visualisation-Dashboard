```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
```

```python
print("HELLO WORLD")
df = pd.read_csv("10_years_network_attack_dataset.csv")


print(df.head())
print(df.info())
print(df.describe())


df['timestamp'] = pd.to_datetime(df['timestamp'])
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month



attack_counts = df['attack_type'].value_counts()
print(attack_counts)
print("Dummy data inserted successfully!")
```

```python

plt.figure()
sns.countplot(data=df, x='attack_type')
plt.xticks(rotation=45)
plt.title("Attack Type Distribution")
plt.show()
print("Visualization Report")
```

```python

plt.figure()
sns.countplot(data=df, x='severity')
plt.title("Severity Level Distribution")
plt.show()
print("Visualization Report")
```

```python


yearly_attacks = df.groupby('year').size()
plt.figure()
plt.plot(yearly_attacks.index, yearly_attacks.values)
plt.title("Year-wise Network Attacks")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
print("Visualization Report")
```

```python

plt.figure()
sns.boxplot(data=df, x='severity', y='anomaly_score')
plt.title("Anomaly Score vs Severity")
plt.show()
print("Visualization Report")
```

```python


plt.figure()
sns.countplot(data=df, x='protocol')
plt.title("Protocol Distribution")
plt.show()
print("Visualization Report")
```

```python

# High Risk Threat Detection

high_risk = df[df['anomaly_score'] > 0.8]
print("High Risk Events:", len(high_risk))

# AI Prompting Simulation

def ai_prompt(prompt, data):
    if "high risk" in prompt.lower():
        return data[data['anomaly_score'] > 0.8]['attack_type'].value_counts()
    if "ddos" in prompt.lower():
        return data[data['attack_type'] == 'DDoS'].groupby('year').size()
    return "Prompt not recognized"

print(ai_prompt("Show high risk attacks", df))
print("END")


```
